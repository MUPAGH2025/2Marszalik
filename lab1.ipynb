```python
import pint
import numpy as np
from matplotlib import pyplot 
from scipy import constants
from open_atmos_jupyter_utils import show_plot
si = pint.UnitRegistry()
si.setup_matplotlib()

2*si.kg + 3*si.mg

R_str = constants.R*si.J/si.K/si.mol 
M_d = 28.966*si.kg/si.kmol
R= R_str/M_d

```

```python
def rho(p,T):       
    return p/R/T
```

```python
rho(1000*si.hPa , 300*si.K).to(si.g/si.cm**3)
```

```python
pyplot.plot(T:= np.linspace(280,320)*si.K , rho(1000*si.hPa, T).to_base_units(), label = "gęstość powietra")
pyplot.gca().xaxis.set_units(si.celsius)
pyplot.legend()
```
